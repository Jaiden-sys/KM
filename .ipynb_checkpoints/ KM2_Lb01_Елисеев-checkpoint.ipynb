**Этап 3. Определение функции перемещения и
 нахождение угла запуска тела**
 Полагаем, что начало прямоугольной декартовой системы координат соответствует уровню земли. Тогда в начальный момент движения тела $t = 0$ имеем:
$$s_x(0) = 0, \quad s_y(0) = h_{start}.$$

Для построения функции перемещения $(s_x(t), s_y(t))$ воспользуемся вторым законом Ньютона:
$$ma = \mathbf{F},$$
где $m$ — масса тела, ускорение движения тела $\mathbf{a} = \bigl(s''_x(t), s''_y(t)\bigr)$ и приложенная сила $\mathbf{F} = m(0, -g)$. В результате имеем два уравнения:
$$s''_x(t) = 0, \quad s''_y(t) = -g.$$

На основании уравнений с учётом начальных условий можно записать аналитический вид для функций $s_x(t)$ и $s_y(t)$:


$$s_x(t) = s_x(0) + v_{0x} t = 0 + v_{0x} t = v_{0x} t,$$

$$s_y(t) = s_y(0) + v_{0y} t - \frac{g}{2} t^2 = h_{start} + v_{0y} t - \frac{g}{2} t^2,$$

где $\mathbf{v_0} = (v_{0x}, v_{0y})$ обозначает неизвестную скорость движения тела в начальный момент времени $t = 0$.

$\color{red}\text{Напишите}$ подробно, как из второго закона Ньютона и начальных
 условий получен аналитический вид функции перемещения 
$(s_x(t), s_y(t))$
 Объяснения оформите в тексте документа с лабораторной работой.

 Аналитический вид функций перемещения получен интегрированием уравнений второго закона Ньютона:  

$$
s''_x(t)=0 \quad \text{и} \quad s''_y(t)=-g
$$  

с учётом начальных условий:  

$$
s_x(0)=0, \quad s_y(0)=h_{start}.
$$  

В результате получаем:  

$$
s_x(t)=v_{0x}t, \quad s_y(t)=h_{start}+v_{0y}t-\frac{g}{2}t^2.
$$

```python
 v0_y = (h_end-h_start+g/2*T**2)/T
 v0_y
```

## Задание 1.2. Определение момента времени касания телом земли 
Выполнение задания осуществляется на основе двух подходов.  

### *Подход 1* (**точное решение**):  
Время касания телом земли находим как точное решение уравнения:  

$$
s_y(t) = 0.
$$  

### *Подход 2* (**приближенное решение**):  

- Постройте последовательность $s_y$ на временном отрезке $[0, t_{stop}]$, чтобы момент касания тела земли $t^*$ попадал в выбранный отрезок, т.е.  
$$
t^* \in [0, t_{stop}].
$$  
Отметим, что $t_{stop} > T$, где значение $T$ задается в Задании 1.1.

- Найдите в построенной последовательности $s_y$ два соседних элемента последовательности, значения которых меняют знак.  
То есть найдите значение индекса $k$, для которого выполняется условие:  
$$
s_y[k] \cdot s_y[k + 1] < 0.
$$  
Это означает, что:  
$$
0 \in \bigl(s_y[k], \, s_y[k + 1]\bigr).
$$  

- Постройте прямую линию $s_{lin}(t)$, проходящую через две точки:  
$$
\bigl(t[k], s_y[k]\bigr) \quad \text{и} \quad \bigl(t[k + 1], s_y[k + 1]\bigr).
$$  

- Найдите значение $t^* \in \bigl(t[k], t[k + 1]\bigr)$ из условия:  
$$
s_{lin}(t^*) = 0.
$$  

Найденное значение $t^*$ будет соответствовать точке $\bigl(t^*, 0\bigr)$ на прямой $s_{lin}(t)$.  

$\color{red}\textbf{Приведите}$ в документе формулы для прямой $s_{lin}(t)$ и для вычисления $t^*$.  



$\color{red}\textbf{Сравните}$ точное решение (*Подход 1*) и приближенное решение (*Подход 2*), вычислив относительную ошибку.

**Подход 1** является точным способом решения при котором аналитически решается уравнение $s_y(t) = 0$, что даёт точное время касания $t^*$ 

**Подход 2** является не таким точным, в нём вычисляется примерное значение при помощи линейной интерполяции между двумя ближайшими точками временной последовательности $s_y$

```python
ar1 = np.arange(0,T+2,0.01)
t1=t2=y1=y2=0
s_y_array1 = h_start+v0_y*ar1-g*ar1**2/2
for k in range(len(s_y_array1) - 1):
    if s_y_array1[k] * s_y_array1[k + 1] < 0:
        t1 = ar1[k]
        y1 = s_y_array1[k]
        t2 = ar1[k+1]
        y2 = s_y_array1[k+1]

k = (y2 - y1) / (t2 - t1)
b = y1 - k * t1
t_zero = -b / k
print("Приближенное значение: ",end=" ")
print(t_zero)
print("Разница между точным и приближенным значением: ", end=" ")
print(t_star_exact - t_zero)
```

Задание 1.4. Построение графиков функций

```python
# Параметры для построения розы
a = 10  # амплитуда
k = 5  # количество лепестков/2 (k ∈ N)
phi = np.linspace(0, 2 * np.pi, 1000)
#создаётся массив равномерно расположенных чисел от 0 до 2*pi размером в 1000 чисел

# Полярное уравнение розы
rho = a * np.sin(2 * k * phi)

# Построение графика в полярной системе координат
plt.figure(figsize=(8, 6))
ax = plt.subplot(1, 2, 1, projection='polar')
ax.plot(phi, rho)
ax.set_title("График розы: ρ = a * sin(2kφ)")

# Параметрические уравнения

t = np.linspace(-5, 5, 1000)
x = np.log(t + np.sqrt(t**2 + 1))
y = t * np.sqrt(t**2 + 1)

# Построение графика в декартовой системе координат
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Параметрический график: x(t), y(t)")
plt.grid()

plt.show()
```
